```python
frames_dir = '/content/drive/MyDrive/code/datasets/OTB/img/Custom'
print(len(os.listdir(frames_dir)))

#make preprocessing
frames_dict = {}
cnt = 0
for image in (os.listdir(frames_dir)):
  img = Image.open(os.path.join(frames_dir,image))
  transforms.ToPILImage(img)
  resize = transforms.Resize((1280,720))
  resized_img = resize(img)
  frames_dict[cnt] = resized_img
  cnt = cnt + 1 

```

```python
# section 2
import cv2

!pip install ipynb
from ipynb.fs.full.part2 import decode_segmap

# define device
# check if there is a GPU available
print(torch.cuda.is_available())
# check what is the current available device
if torch.cuda.is_available():
 print("current device: ", torch.cuda.current_device())
# automatically choose device
dev = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") # use gpu 0 if it is available,
#o.w. use the cpu
print("device: ", dev)

# segmentation using deep method with deep lab v3, same way as in ex. 2.6
dlab = models.segmentation.deeplabv3_resnet101(pretrained=1).eval()
# for a full explaination of the algorithm, see ex 2.5 code
def segment(net, img, show_orig=True , dev='cuda'):
  if show_orig: plt.imshow(img); plt.axis('off'); plt.show()
  # Comment the Resize and CenterCrop for better inference results
  trf = transforms.Compose([transforms.Resize(640), 
                   #T.CenterCrop(224), 
                   transforms.ToTensor(), 
                   transforms.Normalize(mean = [0.485, 0.456, 0.406], 
                               std = [0.229, 0.224, 0.225])])
  inp = trf(img).unsqueeze(0).to(dev)
  out = net.to(dev)(inp)['out']
  max_label = torch.argmax(out.squeeze(), dim=0).detach().cpu().numpy()
  rgb = decode_segmap(max_label)
  return rgb
  #plt.imshow(rgb); plt.axis('off'); plt.show()

segmented_frames = {}
#the prcentage
background_pixels_num = np.zeros(len(frames_dict))
for cnt,image in enumerate(frames_dict):
  segmented_frames[cnt] = segment(dlab, frames_dict.get(image), show_orig=False)
  # check where did the network fail. Too much background pixels mean our object
  # has been sliced.
  background_pixels_num[cnt] = (np.sum(segmented_frames[cnt] == (0,0,0)))
  norm = segmented_frames[0].shape[0]*segmented_frames[0].shape[1]*segmented_frames[0].shape[2]
  background_pixels_num[cnt] = background_pixels_num[cnt]/norm
  #if(background_pixels_num[cnt] > 0.9):
  #  print(cnt)



```

```python
#making dilation in order to compensate some wronged perןpherical segmentations
for cnt,image in enumerate(segmented_frames):
  if(background_pixels_num[cnt] > 0.9):
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(10,100))
    segmented_frames[cnt] = cv2.dilate(segmented_frames[cnt],kernel,iterations = 2)
  #else:
    #continue
    #kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(10,10))
  #plt.imshow(segmented_frames.get(cnt)); plt.axis('off'); plt.show()


#segmenting out
DJ_seg_dict = {}
for cnt,image in enumerate(frames_dict):
  #segmented_frames[cnt] = segment(dlab, frames_dict.get(image), show_orig=False)
  original_img  = np.array(frames_dict.get(cnt))
  seg_img = segmented_frames.get(cnt)
  
  seg_img = cv2.resize(seg_img, (720,1280), interpolation = cv2.INTER_AREA)
  #print(seg_img.shape,cnt,type(seg_img))
  new_img = original_img
  new_img[seg_img == (0,0,0)] = seg_img[seg_img == (0,0,0)]
  #new_img[seg_img != (0,0,0)] = original_img[seg_img != (0,0,0)]
  #plt.imshow(new_img); plt.axis('off'); plt.show()
  DJ_seg_dict[cnt] = new_img


plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.imshow(DJ_seg_dict.get(65))

plt.subplot(1,2,2)
plt.imshow(DJ_seg_dict.get(65))
```

```python
trans = transforms.ToPILImage()
tensor_dict = {}
cutted_dancing_man = {}

for i in deep_method_dict:
  tensor_dict[i] = torch.from_numpy(deep_method_dict[i])
# create a mask
for i in tensor_dict:
  tensor_dict[i] = torch.from_numpy(deep_method_dict[i])
  #print(deep_method_dict.shape)
  mask = torch.zeros_like(tensor_dict[i]).float().to(dev)
  # what labels were recognized?
  label = np.where(deep_method_dict[i])
  #print(label)
  mask[deep_method_dict[i] != 0] = 1 
  input_image =  PIL_dict[i]
  mask = (mask.byte().cpu().numpy())
  mask_PIL = trans(mask)
  mask_PIL = mask_PIL.resize((input_image.size[0],input_image.size[1]))
  mask_arr = np.array(mask_PIL)
  masked_img = input_image * mask_arr
  cutted_dancing_man[i] = masked_img
  # we'll make erosion in order to delete some residual of green background
  # we'll use a tiny kernel in order to keep the visibilty of the man.
  kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
  cutted_dancing_man[i] = cv2.erode(masked_img,kernel,iterations = 1)
  #plt.imshow(cutted_dancing_man[i]); plt.axis('off'); plt.show()

```

```python
#section 4


#function moves the object according to moveX ,moveY
#input: main background, pics from sections: 2,3 , steps to x,y directions, factor for scaled object
#output: image cotains the the distracted object
def localize(mainImage,background,segmented_pic,moveX,moveY,factor = 1):
  scaled_img = mainImage
  scaled_img[0 == segmented_pic] = 0
  # Get extent of interesting part, i.e. non-white part
  y, x, _ = np.nonzero(scaled_img)
  y0, y1 = np.min(y), np.max(y)    # top and bottom rows
  x0, x1 = np.min(x), np.max(x)    # left and right cols
  h, w = y1-y0, x1-x0              # height and width
  ROI = scaled_img[y0:y1, x0:x1]         # extract ROI
  #plt.imshow(segmented_pic); plt.axis('off'); plt.show()

  # Downscale ROI
  scaledROI = cv2.resize(ROI, (w//factor,h//factor), interpolation=cv2.INTER_NEAREST)
  newH, newW = scaledROI.shape[:2]

  # Clear original image to white
  scaled_img[:] = 0
  
  # Get centre of original shape, and position of top-left of ROI in output image
  cx, cy = (x0 + x1) //2, (y0 + y1)//2
  top  = cy - newH//2 + moveY 
  left = cx - newW//2 + moveX
  # Paste in rescaled ROI
  #print(segmented_pic[top:top+newH, left:left+newW].shape,ROI.shape)
  scaled_img[top:top+newH, left:left+newW] = scaledROI
  scaled_img[scaled_img == 0] = background[segmented_pic == 0]

  return scaled_img

########################

#def shifting(mainImage,segmented_pic,moveX,moveY):


# background according to dancing man shape
orig_background = Image.open('/content/drive/MyDrive/data/my_data/original_background_q3.jpg')

orig_background_resized = orig_background.resize((cutted_dancing_man[0].shape[1],cutted_dancing_man[0].shape[0]))
orig_background_arr = np.array(orig_background_resized)

final_frames_dict = {}
print((orig_background_arr.shape))
cnt = 0 #private counter for running dancer's frames
#transplanting the two images
for i in new_DJ_seg_dict:
  DJ_seg_dict_PIL = trans(new_DJ_seg_dict[i])
  DJ_seg_resize = DJ_seg_dict_PIL.resize((cutted_dancing_man[0].shape[1],cutted_dancing_man[0].shape[0]))
  DJ_seg_arr = np.array(DJ_seg_resize)
  dance_man_init = cutted_dancing_man[cnt]
  #plt.imshow(dance_man_init); plt.axis('off'); plt.show()
  dance_man = cutted_dancing_man[i]
  #print(dance_man_init.shape)
  dance_man[dance_man == 0] = orig_background_arr[dance_man == 0]
  dance_man[DJ_seg_arr != 0] = DJ_seg_arr[DJ_seg_arr != 0]
  #plt.imshow(dance_man); plt.axis('off'); plt.show()
  #dance_man = localize(dance_man,orig_background_arr,dance_man_init,-100,20)
  final_frames_dict[i] = dance_man
  cnt = cnt + 1
  
  

```
